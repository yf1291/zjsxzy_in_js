```python
# fname = u'./data/GDP().xlsx'
# fname = u'./data/().xlsx'
# fname = u'./data/().xlsx'
# fname = u'./data/().xlsx'
fname = u'./data/().xlsx'
# fname = u'./data/().xlsx'
# fname = u'./data/().xlsx'
# fname = u'./data/macro.csv'
# fname = u'./data/month_data.xlsx'
mdf = pd.read_excel(fname, index_col=0)
mdf = mdf.fillna(method='ffill').dropna()
fname = u'./data/quarter_data.xlsx'
qdf = pd.read_excel(fname, index_col=0)
```

```python
# df = pd.read_excel(fname, index_col=0)
```

```python
# df = qdf.copy()
# df = df.fillna(method='ffill').dropna(axis=1)
# df = mdf.join(qdf).fillna(method='ffill').dropna()
df = mdf.join(qdf).dropna()
```

```python
# X = df.copy()
X = df.pct_change(periods=12) # /
X = X.replace([-np.inf, np.inf], np.nan).fillna(method='ffill').dropna()
```

```python
X.tail()
```

```python
X.shape
```

# 

```python
ret_df = stock.resample('M').last().pct_change()
```

```python
X = X[X.index >= ret_df.index[0]]
```

```python
ret_df = ret_df.loc[X.index]
```

```python
y = ret_df['000001.SH']
```

```python
X_train, X_test = X[X.index < '2010-01-01'], X[X.index >= '2010-01-01']
y_train, y_test = y[y.index < '2010-01-01'], y[y.index >= '2010-01-01']
```

```python
coef = mutual_info_regression(X, y)
print coef
```

```python
X_train.columns[coef != 0]
```

```python
# reg = ExtraTreesRegressor(max_depth=3)
reg = RandomForestRegressor(max_depth=2)
reg.fit(X, y)
print reg.feature_importances_
# print reg.score(X_train[X_train.columns[coef != 0]], y_train)
# print reg.score(X_test[X_test.columns[coef != 0]], y_test)
```

```python
X.columns[reg.feature_importances_ != 0]
```

```python
plt.figure(figsize=(16, 9))
plt.plot(y_train)
plt.plot(pd.Series(reg.predict(X_train[X_train.columns[coef != 0]]), index=y_train.index))
# print r2_score(reg.predict(X_train), y_train)
# print mean_squared_error(reg.predict(X_train), y_train)
```

```python
y_pred = reg.predict(X_test[X_test.columns[coef != 0]])
plt.figure(figsize=(16, 9))
plt.scatter(y_pred, y_test)
```

```python
y_pred = reg.predict(X_test[X_test.columns[coef != 0]])
plt.figure(figsize=(16, 9))
plt.plot(y_test, label='true')
# plt.plot((1+y_test).cumprod(), label='true')
plt.plot(pd.Series(y_pred, index=y_test.index), label='predict')
# plt.plot((1+pd.Series(y_pred, index=y_test.index)).cumprod(), label='predict')
plt.legend()
# y_test.corr(pd.Series(y_pred, index=y_test.index))
print r2_score(y_pred, y_test)
# print mean_squared_error(y_pred, y_test)
```
