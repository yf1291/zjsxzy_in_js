```python
import pandas as pd
import numpy as np
from WindPy import w
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
w.start()
```

```python
data = w.wsd('USDCNY.FX,EURUSD.FX', 'close', '2010-01-01', '2020-07-09')
```

```python
df = pd.DataFrame(data.Data, columns=data.Times, index=data.Codes)
df = df.T
```

```python
df['USDEUR'] = 1. / df['EURUSD.FX']
```

```python
corr = df['USDCNY.FX'].pct_change().rolling(243).corr(df['USDEUR'].pct_change())
```

```python
fig, ax = plt.subplots(figsize=(16, 9))

ax.plot(corr, color='blue')
```
