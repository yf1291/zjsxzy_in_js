```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
%matplotlib inline
```

```python
WORD_CNT = 'D:\workspace\data\words\wst_word_count.pkl'
TOT_WORD_CNT = 'D:\workspace\data\words\wst_total_word_count.pkl'
```

```python
word_cnt = pd.read_pickle(WORD_CNT)
```

```python
tot_word_cnt = pd.read_pickle(TOT_WORD_CNT)
```

```python
type(word_cnt)
```

```python
word = u'中国'
series = pd.Series(word_cnt[word])
series.index = pd.to_datetime(series.index)
series = series.sort_index()
```

```python
series.plot(figsize=(16, 9))
```

```python
tot = pd.Series(tot_word_cnt)
tot.index = pd.to_datetime(tot.index)
```

```python
tot.plot(figsize=(16, 9))
```

```python
df = pd.read_excel(u'../data/通缩词群.xlsx', index_col=0)
```

```python
df
```

```python
cnt_df = pd.DataFrame()
for word in df.index:
    series = pd.Series(word_cnt[word])
    series.index = pd.to_datetime(series.index)
    series = series.sort_index()
    series = series * 1e5 / tot.loc[series.index]
#     series = pd.Series(preprocessing.scale(series), index=series.index)
    cnt_df[word] = series
```

```python
# cnt_df = cnt_df.div(tot, axis=0)
```

```python
fig = plt.figure(figsize=(16, 9))
plt.bar(cnt_df.index, cnt_df[u'宽松'], width=5)
```

```python
cnt_df = cnt_df.fillna(0)
```

```python
fig = plt.figure(figsize=(16, 9))
plt.bar(cnt_df.index, cnt_df.mul(df['weight']).sum(axis=1), width=5)
```
