```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from adjustText import adjust_text
from WindPy import w
%matplotlib inline
```

```python
w.start()
```

```python
data = w.tdays('2009-12-31', '2020-06-30', "Days=Alldays;Period=S")
```

```python
days = pd.to_datetime(data.Data[0])
```

```python
def get_half_year_return(code, days):
    df = pd.DataFrame(index=days, columns=['ret', 'vol', 'date'])
    for s, e in zip(days, days[1:]):
        data = w.wss(code, "risk_returnyearly_index,stdevry",
                           "startDate=%s;endDate=%s;returnType=1;period=1"%(days[0], e))
        df.loc[e, 'ret'] = data.Data[0][0]
        df.loc[e, 'vol'] = data.Data[1][0]
        df.loc[e, 'date'] = e.strftime('%Y-%m-%d')
    return df
```

```python
tr1 = get_half_year_return('801192.SI', days)
```

```python
tr2 = get_half_year_return('801081.SI', days)
```

```python
tr3 = get_half_year_return('801124.SI', days)
```

```python
fig, ax = plt.subplots(figsize=(16, 9))

ax.plot(tr1['vol'], tr1['ret'], marker='D', color='blue', label='银行')
ax.plot(tr2['vol'], tr2['ret'], marker='D', color='red', label='半导体')
ax.plot(tr3['vol'], tr3['ret'], marker='D', color='darkorange', label='食品加工')

texts = [ax.text(tr1['vol'][i], tr1['ret'][i], tr1['date'][i], fontsize=15) for i in range(1, tr1.shape[0], tr1.shape[0]-2)] + \
        [ax.text(tr2['vol'][i], tr2['ret'][i], tr2['date'][i], fontsize=15) for i in range(1, tr2.shape[0], tr2.shape[0]-2)] + \
        [ax.text(tr3['vol'][i], tr3['ret'][i], tr3['date'][i], fontsize=15) for i in range(1, tr3.shape[0], tr3.shape[0]-2)]
adjust_text(texts)

ax.legend(fontsize=15)
ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
ax.set_ylabel('年化收益率（%）', fontsize=15)
ax.set_xlabel('年化波动率（%）', fontsize=15)
```
