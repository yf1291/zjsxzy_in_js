```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from WindPy import w
%matplotlib inline
```

```python
w.start()
```

```python
data = w.wsd('CJ000005.WI,CJ000004.WI,000906.SH,851114.SI', 'close', '2010-01-01', '2020-11-12')
```

```python
df = pd.DataFrame(data.Data, index=data.Codes, columns=pd.to_datetime(data.Times))
```

```python
df = df.T
```

```python
rets = df.pct_change()
```

```python
rel_rets = pd.DataFrame()
for col in df.columns:
    if col == '000906.SH':
        continue
    else:
        rel_rets[col] = rets[col] - rets['000906.SH']
```

```python
(1 + rel_rets).cumprod().plot(figsize=(16, 9))
```

```python
rel_rets['851114.SI'].corr(rel_rets['CJ000004.WI'])
```

```python
rel_rets['851114.SI'].corr(rel_rets['CJ000005.WI'])
```
