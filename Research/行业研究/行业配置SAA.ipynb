```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from WindPy import w
from adjustText import adjust_text
import empyrical
%matplotlib inline
```

```python
w.start()
```

```python
def metric(daily_ret, period='daily'):
    an_ret = empyrical.annual_return(daily_ret, period=period)
    sharpe = empyrical.sharpe_ratio(daily_ret, risk_free=0.015/365, period=period)
    an_vol = empyrical.annual_volatility(daily_ret, period=period)
    maxdraw = empyrical.max_drawdown(daily_ret)
    print('Annual return: %.2f%%'%(an_ret*100))
    print('Sharpe ratio: %.2f'%(sharpe))
    print('Annual volatility: %.2f%%'%(vol*100))
    print('Max drawdown: %.2f%%'%(maxdraw*100))
    return an_ret, an_vol
```

```python
df = pd.read_excel('./data/中证800行业指数.xlsx', sheet_name='板块基准指数', index_col=0)
```

```python
df.index = pd.to_datetime(df.index, format='%Y%m%d')
```

```python
df = df[df.index >= '2010-01-01']
```

```python
df.iloc[0] = 0
```

```python
df = df[['金融地产', '消费', '医药', '科技', '周期', '制造']]
```

```python
weight = np.array([0.2, 0.3, 0.2, 0.3, 0, 0])
```

```python
def rebalance(df, weight):
    weight_df = pd.DataFrame(index=df.index, columns=df.columns)
    wb = weight
    rebalance_month = set()
    for date in df.index:
        weight_df.loc[date] = wb
        wb = wb * (1 + df.loc[date])
        if not date.strftime('%Y%m') in rebalance_month:
            wb = weight
            rebalance_month.add(date.strftime('%Y%m'))
    pf_ret = (df * weight_df).sum(axis=1)
    return pf_ret
```

```python
# pf_ret = df.mul(weight, axis=1).sum(axis=1)
pf_ret = rebalance(df, weight)
```

```python
an_ret, _ = metric(pf_ret)
```

```python
data = w.wsd('H00906.CSI', 'close', df.index[0], df.index[-1])
```

```python
bench = pd.DataFrame(data.Data, columns=data.Times, index=data.Codes)
bench = bench.T
```

```python
bench_ret = bench.pct_change()
```

```python
bench_an_ret, _ = metric(bench_ret['H00906.CSI'])
```

```python
fig, ax = plt.subplots(figsize=(16, 9))

ax.plot((1 + pf_ret).cumprod() - 1, color='blue', label='行业配置SAA：年化收益率：{:2.2f}%'.format(an_ret*100))
ax.plot((1 + bench_ret['H00906.CSI']).cumprod() - 1, color='red', label='中证800全收益：年化收益率：{:2.2f}%'.format(bench_an_ret*100))
ax.fill_between(pf_ret.index, 0, (1 + pf_ret - bench_ret['H00906.CSI']).cumprod() - 1, color='darkorange', alpha=0.3, label='超额收益率：{:2.2f}%'.format((an_ret-bench_an_ret)*100))

ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
ax.set_ylabel('累计收益率', fontsize=15)
vals = ax.get_yticks()
ax.set_yticklabels(['{:2.0f}%'.format(x*100) for x in vals])
ax.legend(fontsize=15)
```

```python
x = np.arange(0.1, 0.5, 0.1)
y = np.arange(0.1, 0.5, 0.1)
ret, vol, weights, dis = [], [], [], []
for xi in x:
    w1, w2 = xi, 0.5 - xi
    for yi in y:
        w3, w4 = yi, 0.5 - yi
        weight = np.array([w1, w2, w3, w4, 0, 0])
#         pf_ret = df.mul(weight, axis=1).sum(axis=1)
        pf_ret = rebalance(df, weight)
        an_ret, an_vol = metric(pf_ret)
        _, d = metric(pf_ret - bench_ret['H00906.CSI'].fillna(0))
        
        weights.append(('{:0.0f}%'.format(w1*100), '{:0.0f}%'.format(w2*100), '{:0.0f}%'.format(w3*100), '{:0.0f}%'.format(w4*100)))
        ret.append(an_ret)
        vol.append(an_vol)
        dis.append(d)
```

```python
stats = pd.DataFrame({'收益率': ret, '波动率': vol, '权重': weights, '偏离': dis})
```
