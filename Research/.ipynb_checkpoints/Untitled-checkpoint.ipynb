```python
import pandas as pd
import empyrical
from WindPy import w
```

```python
w.start()
```

```python
data = w.wsd('881001.wi', 'close', '2000-01-01', '2020-12-01')
```

```python
df = pd.DataFrame(data.Data, columns=data.Times, index=data.Codes)
df = df.T
```

```python
vol = df['881001.WI'].pct_change().rolling(234).apply(lambda x : empyrical.annual_volatility(x))
```

```python
vol.plot(figsize=(16, 9))
```
