```python
import pandas as pd
import matplotlib.pyplot as plt
from WindPy import w
%matplotlib inline
```

```python
w.start()
```

```python
data = w.wsd('881001.WI', 'close', '2006-01-01', '2018-07-12')
```

```python
pdf = pd.DataFrame(data.Data[0], index=data.Times, columns=['881001.WI'])
```

```python
cons = pd.read_excel('D:/workspace/zjsxzy_in_js/website/everyday-update/data/consistency.xlsx', index_col=0)
```

```python
cons = cons[cons.index >= '2005-01-01']
```

```python
cons['z'] = cons['con60'].rolling(window=243).apply(lambda x: (x[-1] - x.mean()) / x.std())
```

```python
fig = plt.figure(figsize=(16, 9))
ax8 = fig.add_subplot(1, 1, 1)
ax8.plot(cons.index, cons['z'], c='red')
# ax8.set_title(u'市场方向一致度', fontsize=15)
ax88 = ax8.twinx()
ax88.plot(pdf.index, pdf['881001.WI'], color='blue', label=u'万得全A')
ax88.legend(loc='upper right')
```
