```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib
```

```python
df = pd.read_excel(u'./涨价.xlsx', index_col=0)
```

```python
df = df.dropna()
```

```python
df.index = pd.to_datetime(df.index)
```

```python
df.tail()
```

```python
df['mean'] = df['value'].rolling(window=30).mean()
```

```python
df['z'] = df['mean'].rolling(window=500).apply(lambda x: (x[-1] - x.mean()) / x.std())
```

```python
df['z'].plot()
```

```python
cpi = pd.read_excel('./CPI.xls')
```

```python
cpi.tail()
```

```python
cpi = cpi[cpi.index >= '2010-01-01']
df = df[df.index >= '2010-01-01']
```

```python
mdf = df.resample('M').mean()
```

```python
# df = df[(df.index >= '2013-01-01') & (df.index <= '2017-01-01')]
# cpi = cpi[(cpi.index >= '2013-01-01') & (cpi.index <= '2017-01-01')]
```

```python
fig = plt.figure(figsize=(16, 9))
ax = fig.add_subplot(1, 1, 1)
ax.plot(mdf.index, mdf['value'], color='red', label=u'涨价')
# ax.bar(df.index, df['mean'], color='red', width=3, label=u'词频')
ax.legend(loc='upper left')

ax1 = ax.twinx()
ax1.plot(cpi.index, cpi['CPI'], color='blue', label='CPI')
# ax1.plot(cpi.index, cpi['PPI'], color='green', label='PPI')
ax1.legend()
```
