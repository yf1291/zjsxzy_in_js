```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel(u'./中债国债到期收益率10年.xls')
```

```python
df.head()
```

```python
wd = pd.read_excel(u'D:/Data/wei_index/人民币_zt.xlsx', index_col=0)
```

```python
wd.head()
```

```python
df = df.merge(wd, left_index=True, right_index=True)
```

```python
df.index = pd.to_datetime(df.index)
```

```python
df.head()
```

```python
df = df[df.index >= '2014-01-01']
```

```python
fig = plt.figure(figsize=(16, 9))
ax = fig.add_subplot(1, 1, 1)
# ax.bar(df.index, df['value'], color='red', width=3, label=u'词频')
ax.plot(df.index, df['value'], color='red', label=u'词频')
ax.legend(loc='upper left')

ax1 = ax.twinx()
ax1.plot(df.index, df[u'中债国债到期收益率:10年'], color='blue', label=u'10年期国债收益率')
ax1.legend()
```

```python
# df.loc[df[u'衰退'] < df[u'衰退'].mean() + 1.5*df[u'衰退'].std(), u'衰退'] = 0
df.loc[df['value'] < df['value'].mean() + 1.5*df['value'].std(), 'value'] = 0
```

```python
fig = plt.figure(figsize=(16, 9))
ax = fig.add_subplot(1, 1, 1)
ax.bar(df.index, df['value'], color='red', width=3, label=u'词频')
ax.legend(loc='upper left')

ax1 = ax.twinx()
ax1.plot(df.index, df[u'中债国债到期收益率:10年'], color='blue', label=u'10年期国债收益率')
ax1.legend()
```
