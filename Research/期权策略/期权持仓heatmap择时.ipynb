```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import empyrical
from WindPy import w
import os
%matplotlib inline
```

```python
pd.options.mode.chained_assignment = None
DATA_DIR = 'D:/Data/derivatives/options/'
```

```python
files = [f for f in os.listdir(DATA_DIR)]
dic = {}
for f in files:
    fname = '%s/%s'%(DATA_DIR, f)
#     print fname
    option_code = f.rstrip('.xlsx')
    tdf = pd.read_excel(fname, index_col=0)
    dic[option_code] = tdf['oi']
oi_df = pd.DataFrame(dic)
```

```python
w.start()
```

```python
def wind2df(raw_data):
    dic = {}
    for data, field in zip(raw_data.Data, raw_data.Fields):
        dic[field.lower()] = data
    if len(raw_data.Times) == len(raw_data.Data[0]):
        return pd.DataFrame(dic, index=raw_data.Times)
    else:
        return pd.DataFrame(dic)
```

```python
start_date = '2015-02-09'
end_date = '2017-11-30'
```

```python
raw_data = w.wsd("510050.SH", "close", start_date, end_date)
```

```python
df = wind2df(raw_data)
```

```python
df.columns = ['510050.SH']
```

```python
plt.figure(figsize=(16, 9))
df['510050.SH'].plot(c='b')
```

```python
opt_df = pd.read_excel(u'D:/Data/list/.xlsx', index_col=0)
opt_df = opt_df[opt_df['call_or_put'] == u'']
```

```python
opt_df.head()
```

```python
df['opt_price'] = 0.
for date in df.index:
    current = opt_df[(opt_df['listed_date'] <= date) & (opt_df['expire_date'] >= date)]
#     current['oi'] = 0.
    option_code = current.index.map(lambda x: '%d.SH'%(x))
    current['oi'] = oi_df.loc[date].loc[option_code].values
    df.loc[date]['opt_price'] = current[['exercise_price', 'oi']].groupby('exercise_price').sum().idxmax()['oi']
```

```python
plt.figure(figsize=(16, 9))
df['510050.SH'].plot(label='50ETF', c='b')
df['opt_price'].plot(label='max volume price of put options')
plt.legend()
```

```python
plt.figure(figsize=(16, 9))
(df['510050.SH'] - df['opt_price']).plot(c='b', label='spread')
(df['510050.SH'] - 2).plot(label='50ETF')
pd.Series(0, index=df.index).plot(label='')
plt.legend()
```
