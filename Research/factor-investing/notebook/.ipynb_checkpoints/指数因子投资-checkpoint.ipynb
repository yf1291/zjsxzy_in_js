### 相关性均值的变化

```python
def roll(df, w):
    """
    This fucntion comes from:
    http://stackoverflow.com/questions/37486502/why-does-pandas-rolling-use-single-dimension-ndarray/37491779#37491779
    """
    df.fillna(df.mean(), inplace=True)
    roll_array = np.dstack([df.values[i:i+w, :] for i in range(len(df.index) - w + 1)]).T
    panel = pd.Panel(roll_array,
                     items=df.index[w-1:],
                     major_axis=df.columns,
                     minor_axis=pd.Index(range(w), name='roll'))
    return panel.to_frame().unstack().T.groupby(level=0)
```

```python
rolled_df = roll(factor_ret_df, 121)
```

```python
corr_series = rolled_df.apply(lambda x: x.corr().mean().mean())
```

```python
plt.figure(figsize=(16, 9))
plt.plot(corr_series)
```

# 因子资产配置（月度调仓）

```python
pnl.ix[:, :, 'momentum per'] = pnl.minor_xs('momentum').rolling(window=484).apply(lambda x: utils.rank_percentile(x))
```

```python
pnl.ix[:, :, 'carry per'] = pnl.minor_xs('carry').rolling(window=484).apply(lambda x: utils.rank_percentile(x))
```

```python
pnl.ix[:, :, 'value per'] = pnl.minor_xs('value').rolling(window=484).apply(lambda x: utils.rank_percentile(x))
```

```python
pnl.ix[:, :, 'defensive per'] = pnl.minor_xs('defensive').rolling(window=484).apply(lambda x: utils.rank_percentile(x))
```

```python
rebanlance_dates = pnl.minor_xs('momentum per').dropna().resample('BM').last().index
```

```python
weight_df = pd.DataFrame(columns=pnl.items, index=pnl.minor_xs('momentum per').dropna().index)
```

```python
pnl.items
```

```python
weight = pd.Series(0, index=pnl.items)
k = 3
for date in weight_df.index:
    weight_df.loc[date] = weight
    if date in rebanlance_dates:
        # rebalance
        weight = pd.Series(0, index=pnl.items)
        df = pnl.major_xs(date).T
        
        # carry
        weight[df['carry per'].sort_values(ascending=False)[:k].index] += 1
        # value
        weight[df['value per'].sort_values(ascending=False)[:k].index] += 1
        # defensive
        weight[df['defensive per'].sort_values()[:k].index] += 1
        # momentum
        weight[df['momentum per'].sort_values(ascending=False)[:5].index] += 1
        
        weight /= weight.sum()
        print date
        print weight[weight != 0]
```

```python
ret = (weight_df * pnl.minor_xs('return').loc[weight_df.index]).sum(axis=1)
```

```python
utils.get_metrics(ret)
```

```python
acc_ret = (1 + ret).cumprod()
```

```python
plt.figure(figsize=(16, 9))
plt.plot(acc_ret, label='acc return')
plt.plot((1+pnl.minor_xs('return').loc[ret.index]['000300.SH']).cumprod(), label='000300')
plt.legend()
```
