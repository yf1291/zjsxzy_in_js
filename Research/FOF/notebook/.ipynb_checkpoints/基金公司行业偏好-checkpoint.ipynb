```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
FOF_DIR = 'D:/workspace/data/FOF'
```

```python
df = pd.read_excel('%s/comp_ret.xlsx'%(FOF_DIR), index_col=0)
```

```python
df = (1 + df).cumprod()
```

```python
df.head()
```

```python
DATA_DIR = 'D:/Data/index/'
```

```python
files = [f for f in os.listdir(DATA_DIR) if f.startswith('CI')]
```

```python
dic = {}
for f in files:
    name = f.rstrip('.xlsx')
    temp = pd.read_excel('%s/%s'%(DATA_DIR, f), index_col=0)
    dic[name] = temp['close']
ind_df = pd.DataFrame(dic)
```

```python
ind_df = (1 + ind_df.pct_change()).cumprod()
```

```python
ind_df.dropna().head()
```

```python
cor_df = pd.DataFrame(columns=ind_df.columns, index=df.columns)
```

```python
for ind in ind_df.columns:
    for comp in df.columns:
        sa, sb = ind_df[ind], df[comp]
        sa, sb = sa.dropna(), sb.dropna()
        sa = sa[sa.index >= sb.index[0]]
        sb = sb[sb.index >= sa.index[0]]
#         sa, sb = sa.resample('BM').last(), sb.resample('BM').last()
        cor_df.loc[comp, ind] = sa.corr(sb)
```

```python
cor_df.to_excel('../data/comp_cor.xlsx')
```
