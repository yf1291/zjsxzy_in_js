```python
from WindPy import w
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pyfolio as pf
%matplotlib inline
```

```python
def roll(df, w):
    """
    This fucntion comes from:
    http://stackoverflow.com/questions/37486502/why-does-pandas-rolling-use-single-dimension-ndarray/37491779#37491779
    """
    df.fillna(df.mean(), inplace=True)
    roll_array = np.dstack([df.values[i:i+w, :] for i in range(len(df.index) - w + 1)]).T
    panel = pd.Panel(roll_array,
                     items=df.index[w-1:],
                     major_axis=df.columns,
                     minor_axis=pd.Index(range(w), name='roll'))
    return panel.to_frame().unstack().T.groupby(level=0)
```

```python
fund = '001616.OF'
index = '886035.WI'
```

```python
fund_df = pd.read_excel('%s/%s.xlsx'%(FOF_DIR, fund), index_col=0)
```

```python
fund_df = fund_df[['nav_adj']].dropna()
```

```python
fund_df.head()
```

```python
index_df = pd.read_excel('%s/%s.xlsx'%(INDEX_DIR, index), index_col=0)
```

```python
df = index_df.pct_change().join(fund_df.pct_change()).dropna()
```

```python
df['nav_adj'] = pf.empyrical.cum_returns(df['nav_adj'])
df['close'] = pf.empyrical.cum_returns(df['close'])
```

```python
df.plot()
```

```python
roll_df = roll(df, 60)
```

```python
corr = roll_df.apply(lambda x: x['nav_adj'].corr(x['close']))
```

```python
plt.figure(figsize=(16, 9))
plt.plot(corr)
```
